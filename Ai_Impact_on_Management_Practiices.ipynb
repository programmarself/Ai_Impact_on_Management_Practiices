# 📘 AI Impact on Management Practices - A to Z Analysis

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import os
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
%matplotlib inline
```

```python

df = pd.read_csv("ai_management_data.csv")
df.head()
```

```python

print("Shape:", df.shape)
print("Missing values:\n", df.isnull().sum())
print("Data types:\n", df.dtypes)
print("Unique values:\n", df.nunique())
```

### 📊 Industry Frequency

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Industry', palette='Set3')
plt.title("📊 Industry Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 📊 Department Frequency

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Department', palette='Set2')
plt.title("📊 Department Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 📊 AI Impact Score by Industry

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Industry', y='AI_Impact_Score', palette='coolwarm')
plt.title("📊 AI Impact Score by Industry")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 📊 AI Score by Department

```python
plt.figure(figsize=(10, 5))
sns.violinplot(data=df, x='Department', y='AI_Impact_Score', palette='mako')
plt.title("📊 AI Score by Department")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 📊 Impact Type Distribution

```python
plt.figure(figsize=(10, 5))
df['Impact_Type'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('pastel')); plt.ylabel('')
plt.title("📊 Impact Type Distribution")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 📊 Heatmap of Avg AI Score

```python
plt.figure(figsize=(10, 5))

pivot = df.pivot_table(index="Industry", columns="Department", values="AI_Impact_Score", aggfunc="mean")
sns.heatmap(pivot, annot=True, cmap="YlGnBu", fmt=".1f")

plt.title("📊 Heatmap of Avg AI Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 🌞 Sunburst Chart

```python

fig = px.sunburst(df, path=['Industry', 'Department', 'Impact_Type'],
                  values='AI_Impact_Score', color='AI_Impact_Score',
                  color_continuous_scale='Blues',
                  title="Sunburst Chart: AI Score Hierarchy")
fig.show()
```

### 🧱 Treemap Chart

```python

fig = px.treemap(df, path=['Industry', 'Department', 'Impact_Type'],
                 values='AI_Impact_Score', color='AI_Impact_Score',
                 color_continuous_scale='RdBu',
                 title="Treemap: AI Impact Score")
fig.show()
```

### 🎯 Strip Plot

```python

sns.stripplot(data=df, x="Impact_Type", y="AI_Impact_Score", jitter=True, palette="Set1")
plt.title("AI Impact Score by Impact Type")
plt.tight_layout()
plt.show()
```

### 🔁 Pair Plot

```python

sns.pairplot(df, hue="Impact_Type")
plt.suptitle("Pair Plot", y=1.02)
plt.show()
```
